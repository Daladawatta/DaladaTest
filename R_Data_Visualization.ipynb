## Generate Individual R Plots

### Subtask:
Generate the three R plots (Count vs RSC, 500001-9999999 vs RSC, and 100001-500000 vs RSC) and store them as individual plot objects in R.


**Reasoning**:
The subtask requires generating three R plots as individual objects. I will use the `%%R` magic command to execute R code, `ggplot2` to create the bar plots, assign each plot to a variable, and set appropriate labels and theme settings, including rotating the x-axis labels for readability.



```python
%%R

library(ggplot2)

# Create a bar plot of RSC vs Count
plot_count <- ggplot(r_df, aes(x = RSC, y = Count)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Count by RSC", x = "RSC", y = "Count") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Create a bar plot of RSC vs 500001-9999999
plot_500k_9m <- ggplot(r_df, aes(x = RSC, y = `500001-9999999`)) +
  geom_bar(stat = "identity", fill = "darkgreen") +
  labs(title = "Amount in 500001-9999999 Range by RSC", x = "RSC", y = "Amount (500001-9999999)") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Create a bar plot of RSC vs 100001-500000
plot_100k_500k <- ggplot(r_df, aes(x = RSC, y = `100001-500000`)) +
  geom_bar(stat = "identity", fill = "purple") +
  labs(title = "Amount in 100001-500000 Range by RSC", x = "RSC", y = "Amount (100001-500000)") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

print("R plots created and stored as objects: plot_count, plot_500k_9m, plot_100k_500k")
```

**Reasoning**:
The subtask requires combining the three previously generated R plots (`plot_count`, `plot_500k_9m`, `plot_100k_500k`) into a single dashboard using the `patchwork` library. This involves loading the `patchwork` library, arranging the plots, and then printing the combined plot.



```python
%%R

# Load the patchwork library for combining plots
library(patchwork)

# Combine the plots into a dashboard
combined_plot <- plot_count + plot_500k_9m + plot_100k_500k + plot_layout(ncol = 1)

# Print the combined plot
print(combined_plot)
```
